## Clean up the data
Did you notice anything unusual about the "house" amounts?

How about the handset price and income?

```python
df_clean = df[(df['house'] > 0) & (df['income'] > 0) & (df['handset_price'] < 1000)]
df_clean.describe()
```

# Add Derived Attribute

Add a column with the income in K (thousands)

```python
df_clean['income_k'] = (df_clean['income']/1000).round(2)
df_clean.head()
```

## Summarize the Data
 What is average house price at the two levels of leave?

 Conclusion?  Stayers have substantially more expensive homes.

```python
df_clean[['house','leave']].groupby('leave').mean().round(2)
```

## Data Visualization

These charts show satisfaction levels compared to other factors and trends.

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.countplot(x='reported_satisfaction', data=df_clean)
plt.title('Reported Satisfaction Levels')
plt.show()
```

```python

sns.barplot(x='reported_satisfaction', y='average_call_duration', data=df_clean,)
plt.title('Average Call Duration by Reported Satisfaction')
plt.xlabel('Reported Satisfaction')
plt.ylabel('Average Call Duration')
plt.show()
```
